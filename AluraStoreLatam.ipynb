#1. Análisis de facturación



```python
def suma_x_tienda(td):

  ingreso_t1 = 0
  for i in td['Precio']:
    ingreso_t1 += i
  return ingreso_t1

suma_t1 = suma_x_tienda(tienda)
suma_t2 = suma_x_tienda(tienda2)
suma_t3 = suma_x_tienda(tienda3)
suma_t4 = suma_x_tienda(tienda4)

print('Ingreso total tienda 1: ', int(suma_t1))
print('Ingreso total tienda 2: ', int(suma_t2))
print('Ingreso total tienda 3: ', int(suma_t3))
print('Ingreso total tienda 4: ', int(suma_t4))
```

```python
x = np.array(['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'])
y = np.array([suma_t1, suma_t2, suma_t3, suma_t4])

plt.barh(x, y, color='green', height= 0.5)
plt.xlabel('Ingreso')
plt.ylabel('Tiendas')
plt.show()
```

# 4. Productos más y menos vendidos

```python
def productos_vendidos_x_tienda(td):
  productos_vendidos = []
  for prod in td['Producto']:
    if prod not in productos_vendidos:
      productos_vendidos.append(prod)
  return productos_vendidos

productos_t1 = productos_vendidos_x_tienda(tienda)
productos_t2 = productos_vendidos_x_tienda(tienda2)
productos_t3 = productos_vendidos_x_tienda(tienda3)
productos_t4 = productos_vendidos_x_tienda(tienda4)

def cantidad_productos_vendidos(tienda, productos):

  conteo_productos = {}
  for x in range(len(productos)):
    for prod in tienda['Producto']:
      if prod == productos[x]:
        conteo_productos[productos[x]] = conteo_productos.get(productos[x], 0) + 1
  return conteo_productos

cantidad_productos_vendidos_tienda = cantidad_productos_vendidos(tienda, productos_t1)
cantidad_productos_vendidos_tienda2 = cantidad_productos_vendidos(tienda2, productos_t2)
cantidad_productos_vendidos_tienda3 = cantidad_productos_vendidos(tienda3, productos_t3)
cantidad_productos_vendidos_tienda4 = cantidad_productos_vendidos(tienda4, productos_t4)

def tabla_prd(cntpr):
  df = pd.Series(cntpr, name= "Cantidad").reset_index()
  df.columns = ['Producto', 'Cantidad']

  return df
```

```python
def orden_cantidad_productos(cp):
  orden_desc = sorted(cp.items(), key=lambda x: x[1], reverse=True)
  return orden_desc

orden_cantidad_productos_tienda = orden_cantidad_productos(cantidad_productos_vendidos_tienda)
orden_cantidad_productos_tienda2 = orden_cantidad_productos(cantidad_productos_vendidos_tienda2)
orden_cantidad_productos_tienda3 = orden_cantidad_productos(cantidad_productos_vendidos_tienda3)
orden_cantidad_productos_tienda4 = orden_cantidad_productos(cantidad_productos_vendidos_tienda4)

prod_max, cant_max = orden_cantidad_productos_tienda[0]
prod_max2, cant_max2 = orden_cantidad_productos_tienda2[0]
prod_max3, cant_max3 = orden_cantidad_productos_tienda3[0]
prod_max4, cant_max4 = orden_cantidad_productos_tienda4[0]

prod_min, cant_min = orden_cantidad_productos_tienda[-1]
prod_min2, cant_min2 = orden_cantidad_productos_tienda2[-1]
prod_min3, cant_min3 = orden_cantidad_productos_tienda3[-1]
prod_min4, cant_min4 = orden_cantidad_productos_tienda4[-1]

print(f'En la primera tienda el producto más vendido es: {prod_max} con una cantidad de: {cant_max} unidades')
print(f'En la segunda tienda el producto más vendido es: {prod_max2} con una cantidad de: {cant_max2} unidades')
print(f'En la tercera tienda el producto más vendido es: {prod_max3} con una cantidad de: {cant_max3} unidades')
print(f'En la cuarta tienda el producto más vendido es: {prod_max4} con una cantidad de: {cant_max4} unidades\n')

print(f'En la primera tienda el producto menos vendido es: {prod_min} con una cantidad de: {cant_min} unidades')
print(f'En la segunda tienda el producto menos vendido es: {prod_min2} con una cantidad de: {cant_min2} unidades')
print(f'En la tercera tienda el producto menos vendido es: {prod_min3} con una cantidad de: {cant_min3} unidades')
print(f'En la cuarta tienda el producto menos vendido es: {prod_min4} con una cantidad de: {cant_min4} unidades')
```

```python
def promedio_costo_envio(envio):

  prom = 0
  for env in envio['Costo de envío']:
    prom = prom + env
  prom_envio = prom / len(envio['Costo de envío'])
  return prom_envio

envios_tienda = promedio_costo_envio(tienda)
envios_tienda2 = promedio_costo_envio(tienda2)
envios_tienda3 = promedio_costo_envio(tienda3)
envios_tienda4 = promedio_costo_envio(tienda4)

print(f'Promedio costo de envío tienda 1: {envios_tienda:.2f}')
print(f'Promedio costo de envío tienda 2: {envios_tienda2:.2f}')
print(f'Promedio costo de envío tienda 3: {envios_tienda3:.2f}')
print(f'Promedio costo de envío tienda 4: {envios_tienda4:.2f}')
```
